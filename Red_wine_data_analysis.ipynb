## USING K NEAREST NEIGHBORS ALGORITHM

```python
from sklearn.neighbors import KNeighborsClassifier as knn
mod=knn(n_neighbors=23)
```

```python
X=data.drop(['quality'],axis=1).values
#the data is divided into two parts X and y X is the faetures data and y is the label data both will be used to train and 
#test the model
y=data['quality'].values 
```

```python




X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=42,stratify=y)
# We perform hyperparameter tuning during Train-test split and choose the optimum values to provide maximum accuracy
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python

mod.fit(X_train,y_train)
```

```python
accuracy=mod.score(X_test,y_test)*100
print("the accuracy of the model is",accuracy)
```

# USING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression as LRR
```

```python
logreg = LRR()
```

```python

X=data.drop(['quality'],axis=1)#Feature Variables
y=data['quality']#Target Variable
```

```python
#implementing train-test model
from sklearn.model_selection import train_test_split
# We perform hyperparameter tuning during Train-test split and choose the optimum values to provide maximum accuracy
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=25,stratify=y)
```

```python
logreg.fit(X_train, y_train)
```

```python
accuracy2 = logreg.score(X_test, y_test) * 100
print("Accuracy of the model while using Logistic Regression is " + str(accuracy2))
```
